# Simple Calculator

```c
#include <stdio.h>

int main()
{
    int x, y, sum, sub, multiply;
    float div;
    printf("Enter two numbers: ");
    scanf("%d %d", &x, &y);
    sum = x + y;
    sub = x - y;
    multiply = x * y;
    div = (float)x / y; // at least one value should be float. It could be either x or y. 

    printf("%d + %d = %d\n", x, y, sum);
    printf("%d - %d = %d\n", x, y, sub);
    printf("%d * %d = %d\n", x, y, multiply);
    printf("%d / %d = %.2f\n", x, y, div);
  
    return 0;  
}
```

### OUTPUT

```console
Enter two numbers: 25 25
25 + 25 = 50 
25 - 25 = 0  
25 * 25 = 625
25 / 25 = 1.00
```

# Maximum number among three numbers

```c
// find the maximum number among three numbers. 
// using nested if. 
#include <stdio.h>

int main(){
    int x, y, z;
    printf("Enter three numbers: ");
    scanf("%d %d %d", &x, &y, &z);
    printf("x = %d, y = %d, z = %d\n", x, y, z);

    if (x > y){
        if (x > z){
            printf("Maximum = %d\n", x);
        }
    }

    if (y > x){
        if(y > z){
            printf("Maximum = %d\n", y);
        }
    }

    if ( z > y){
        if(z > x){
            printf("Maximum = %d\n", z);
        }
    }

    return 0;
}
```

### OUTPUT

```console
Enter three numbers: 67 89 25
x = 67, y = 89, z = 25
Maximum = 89
```

# Find if an input is Alphabet or number or special character.

```c
// Find if an input is Alphabet or number or special character.
#include <stdio.h>

int main(){
    char ch;
    printf("Enter a Character: ");
    scanf(" %c", &ch);

    if((ch >= 'a' && ch <= 'z') || (ch >= 'A' && ch <= 'Z')){
        printf("Alphabet\n");
    } else if(ch >= '0' && ch <= '9'){
        printf("Number\n");
    } else{
        printf("Special Character\n");
    }

    return 0;
}
```

### OUTPUT

```console
Enter a Character: u
Alphabet
```

# Proper leap year
